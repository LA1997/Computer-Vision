```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# IMAGE 1

```python
img = cv2.imread('Images/face_boy.jpeg')
plt.imshow(img)
```

```python
# First let us apply canny edge Detection method directly ie without blurring

edges = cv2.Canny(image=img,threshold1=127,threshold2=127)
plt.imshow(edges)
```

```python
# Choosing the threshold Starting points and Blurring the image

med_val = np.median(img)

lower = int(max(0,0.5*med_val))
upper = int(min(255,med_val))

blurred_img = cv2.blur(img,ksize=(2,2))
```

```python
med_val
```

```python
# Now applying canny edge detection method

edges = cv2.Canny(image= blurred_img,threshold1=lower,threshold2=upper)
plt.imshow(edges)
```
