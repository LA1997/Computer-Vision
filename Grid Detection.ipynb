```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('Images/chess.jpg')
print(img)
plt.imshow(img)
```

```python
found,corners = cv2.findChessboardCorners(img,(9,9))
```

```python
if found:
    cv2.drawChessboardCorners(img,(9,9),corners,found)
```
