```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# IMAGE 1

```python
maze = cv2.imread('Images/maze.png')
maze = cv2.cvtColor(maze,cv2.COLOR_BGR2RGB)
plt.imshow(maze)
maze.shape
```

```python
# Convert it to grayScale
gray_maze = cv2.cvtColor(maze,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_maze,cmap = 'gray')
gray_maze.shape
```

```python
gray_maze = np.float32(gray_maze)
```

```python
############# CORNER HARRIS ALGORITHM ##################

dst = cv2.cornerHarris(src=gray_maze,blockSize=2,ksize=3,k=0.04)
dst = cv2.dilate(dst,None)

maze[dst>0.01*dst.max()] = [255,0,0]
plt.imshow(maze)
```

```python
############# Shi Tamasi ALGORITHM ##################
# (Reload the previous cells except above one cell)
corners = cv2.goodFeaturesToTrack(gray_maze,maxCorners=1000,qualityLevel=0.01,minDistance=10)

# Visualize
corners = np.int0(corners)
for i in corners:
    x,y = i.ravel()
    cv2.circle(maze,(x,y),3,(255,0,0),-1)
plt.imshow(maze)
```
